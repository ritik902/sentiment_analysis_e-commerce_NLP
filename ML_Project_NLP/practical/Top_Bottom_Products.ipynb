```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("flipkart_product_cleaned_final_sentiment.csv")
```

```python
df.head()
```

```python
import re
# Function to clean product names
def clean_product_name(name):
    # Remove all characters except letters, numbers, and spaces
    return re.sub(r'[^A-Za-z0-9(),./\s]', '', name)

# Apply the function to the 'ProductName' column
df['CleanedProductName'] = df['ProductName'].apply(clean_product_name)

# Show the first few rows to verify the changes
print(df[['ProductName', 'CleanedProductName']].head())
```

```python
# Group the reviews by ProductName and Rate to count the number of positive and negative reviews for each product
product_sentiment_counts=df.groupby(['CleanedProductName', 'Rate']).size().unstack(fill_value=0)

# Rename the columns for clarity
product_sentiment_counts.columns=['Negative', 'Positive']

# Show the grouped data
product_sentiment_counts.head()
```

```python
product_sentiment_counts.sort_values(by='Positive', ascending=False).head(10)
```

```python
# Sort the products by the number of positive reviews to find the top 10 products
top_10_products = product_sentiment_counts.sort_values(by='Positive', ascending=False).head(10)

# Sort the products by the number of negative reviews to find the bottom 10 products
bottom_10_products = product_sentiment_counts.sort_values(by='Negative', ascending=False).head(10)

# Plot the top 10 products
plt.figure(figsize=(10, 6))
top_10_products.plot(kind='bar', stacked=True, color=['red', 'green'], ax=plt.gca())
plt.title('Top 10 Products with the Most Positive Reviews')
plt.xlabel('Product Name')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=90)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()

# Plot the bottom 10 products
plt.figure(figsize=(10, 6))
bottom_10_products.plot(kind='bar', stacked=True, color=['red', 'green'], ax=plt.gca())
plt.title('Bottom 10 Products with the Most Negative Reviews')
plt.xlabel('Product Name')
plt.ylabel('Number of Reviews')
plt.xticks(rotation=90)
plt.legend(title='Sentiment')
plt.tight_layout()
plt.show()
```
