This function i used to remove extra things from price column i.e. f?? , A.. something like that so i user regular expression for each row

\d is a shorthand character that matches any digit (0-9)... so ^ neglect the \d so it matches anything that is not a digit (0-9)

```python
import re
def clean_price(price):
    clean_price=re.sub(r'[^\d]', '', str(price))
    return clean_price if clean_price else '0'  # Replace empty strings with '0'
df.loc[:,'Price']=df['Price'].apply(clean_price)
```

```python
# 1999 rows are Nan in Review column
# Fill NaN values in the 'Review' column based on 'Rate'
df.loc[(df['Review']=='Nan') & (df['Rate'] == 5), 'Review'] = 'Just Buy'  # Assuming rate > 1 indicates positive sentiment
df.loc[(df['Review']=='Nan') & (df['Rate'] == 4), 'Review'] = 'Highly Recommend'
df.loc[(df['Review']=='Nan') & (df['Rate'] == 3), 'Review'] = 'Good Quality'
df.loc[(df['Review']=='Nan') & (df['Rate'] == 2), 'Review'] = 'Bad Quality'
df.loc[(df['Review']=='Nan') & (df['Rate'] == 1), 'Review'] = 'Dont Consider'  # Assuming rate == 1 indicates negative sentiment
```

Now i remove the stop words from summary column... stop words are common words that do not carry significant meaning and are usually removed to focus on the important words in the text.

```python
import nltk
nltk.download('stopwords')
nltk.download('punkt')
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

This function is used to remove unnecessary characters from summary column and only take alphabets because it is a categorical column and also it convert them to lower case and remove the stop words from summary column

```python
def remove_stop_words(text):
    text=re.sub(r'[^a-zA-Z\s]', '', text) # or --> re.sub(r'[^a-zA-Z\s,]', '', text)
    text=text.lower()
    stop_words=set(stopwords.words('english'))
    word_tokens=word_tokenize(text)
    # filtered_text=[word for word in word_tokens if word.lower() not in stop_words]
    filtered_text = []
    for i in word_tokens:
        if i.lower() not in stop_words:
            filtered_text.append(i)
    return ' '.join(filtered_text)
```

```python
df.loc[:,'Summary']=df['Summary'].astype(str).apply(remove_stop_words)
```

```python
# Stemming and lemmatization
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
# Initialize the stemmer and lemmatizer
stemmer = PorterStemmer()
lemmatizer = WordNetLemmatizer()

# Function for stemming
def stem_text(text):
    tokens = word_tokenize(text)  # Tokenize the text
    stemmed = [stemmer.stem(token) for token in tokens]  # Stem each word
    return ' '.join(stemmed)  # Join the tokens back to a single string

# Function for lemmatization
def lemmatize_text(text):
    tokens = word_tokenize(text)  # Tokenize the text
    lemmatized = [lemmatizer.lemmatize(token) for token in tokens]  # Lemmatize each word
    return ' '.join(lemmatized)  # Join the tokens back to a single string
```

```python
# Apply stemming using loc and create new columns
df.loc[:,'Review']=df['Review'].apply(stem_text)
df.loc[:,'Summary']=df['Summary'].apply(stem_text)

# Apply lemmatization using loc and create new columns
df.loc[:,'Review']=df['Review'].apply(lemmatize_text)
df.loc[:,'Summary']=df['Summary'].apply(lemmatize_text)
```

```python
from textblob import TextBlob
```

Here i use textblob to get the sentiment from summary column 

```python
def sentiment(text):
    return TextBlob(text).sentiment.polarity
df.loc[:,'sentiment']=df['Summary'].apply(sentiment)
```

Sentiment is group by with review,price and rate column and find mean to get the sentiment comparison values

```python
Review_sentiment = df.groupby('Review')['sentiment'].mean().reset_index()
print(Review_sentiment)
```

```python
import seaborn as sns
```

```python
plt.figure(figsize=(20, 5))
sns.barplot(x='Review', y='sentiment', data=Review_sentiment)
plt.title('Average Sentiment by Review')
plt.xlabel('Review')
plt.ylabel('Average Sentiment')
plt.show()
```
